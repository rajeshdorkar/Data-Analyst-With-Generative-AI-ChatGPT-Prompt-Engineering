# AI Greeting Card Assistant (QuickGreetings AI)

This Jupyter Notebook demonstrates how to build a simple Generative AI application using the Gemini API.
The app takes an occasion, recipient, and tone, then generates a custom greeting message.

## Part 1: Install Required Libraries

```python
# Run this cell in Google Colab or your local environment
!pip install streamlit google-generativeai
```

## Part 2: Configure Gemini API Key

```python
import os

# Option 1: Set API key directly (for notebook use only)
# os.environ['GEMINI_API_KEY'] = 'your_api_key_here'

api_key = os.getenv('GEMINI_API_KEY')
if not api_key:
    raise ValueError('GEMINI_API_KEY not found in environment variables')
```

## Part 3: Initialize Gemini Model

```python
import google.generativeai as genai

genai.configure(api_key=api_key)
model = genai.GenerativeModel('gemini-pro')
```

## Part 4: Generate Greeting Card Message

```python
# User inputs
occasion = 'Birthday'
recipient = 'my best friend'
tone = 'Funny'

prompt = f(
    f'Write a {tone.lower()} greeting card message for {recipient} on their {occasion}. '
    'Keep it friendly and engaging.'
)

response = model.generate_content(prompt)
print(response.text)
```
