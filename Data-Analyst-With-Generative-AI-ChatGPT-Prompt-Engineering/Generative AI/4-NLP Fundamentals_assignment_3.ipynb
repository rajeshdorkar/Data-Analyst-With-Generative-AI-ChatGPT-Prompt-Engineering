```python
import nltk
from nltk.tokenize import word_tokenize
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
```

## 1. Tokenization

```python
sentence = "Prompt engineering is the art of guiding AI models effectively."
tokens = word_tokenize(sentence)

print("Tokens:")
print(tokens)
```

## 2. Semantic Similarity

```python
sentences = [
    "The king is the ruler of the country.",
    "The queen governs her people with wisdom.",
    "I love a good pepperoni pizza.",
    "Monarchy is a form of government."
]

model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings
embeddings = model.encode(sentences)

# Compute similarity
base_embedding = embeddings[0].reshape(1, -1)
similarities = cosine_similarity(base_embedding, embeddings[1:])[0]

# Print similarity scores
for i, score in enumerate(similarities, start=1):
    print(f"Sentence: {sentences[i]}")
    print(f"Similarity Score: {score:.4f}\n")

# Most similar sentence
most_similar_index = similarities.argmax() + 1

print("Most similar sentence:")
print(sentences[most_similar_index])
```
