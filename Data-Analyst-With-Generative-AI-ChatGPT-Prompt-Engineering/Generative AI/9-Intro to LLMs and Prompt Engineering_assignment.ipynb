# Assignment 1: Fundamentals – The E‑Commerce Chatbot Assistant

This notebook demonstrates few-shot prompting, temperature control, and contextual question answering for the Gadgetron e‑commerce chatbot.

## Task 1: Ticket Routing with Few‑Shot Prompting

### Prompt Used
```text
You are a support ticket routing assistant for an online electronics store called Gadgetron.
Your job is to classify customer emails into: Shipping, Billing, or Returns.

Email: My order hasn’t arrived yet and tracking hasn’t updated.
Department: Shipping

Email: I was charged twice for my purchase.
Department: Billing

Email: I want to return a product I don’t need anymore.
Department: Returns

Email: Hi, I received my order (#GT-58291) today, but the headphones are broken. How do I send them back for a replacement?
Department:
```

**Model Output:** Returns

## Task 2: Generating Responses with Temperature

### Low Temperature (Standard & Professional)
```text
Write a standard and professional apology for a delayed order.
```

**Model Output:**
We apologize for the delay in delivering your order. Please accept our sincere apologies for any inconvenience caused.

### High Temperature (Empathetic & Creative)
```text
Write an empathetic and creative apology for a delayed order.
```

**Model Output:**
We’re truly sorry your package took the scenic route to your doorstep. We know waiting is frustrating, and we really appreciate your patience!

## Task 3: Answering Questions with Context

### Prompt with Context
```text
Return Policy:
Customers can return unused items within 30 days for a full refund.
Customers pay return shipping. Sale items are non-refundable.

Question: Can I return a laptop I bought on sale last week? Will you pay for shipping?
```

**Model Output:**
Sale items are non-refundable, and customers are responsible for return shipping costs.
