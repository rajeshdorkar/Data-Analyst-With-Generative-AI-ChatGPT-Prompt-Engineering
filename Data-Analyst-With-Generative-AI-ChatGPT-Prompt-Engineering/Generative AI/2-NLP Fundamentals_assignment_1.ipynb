```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

## Sample Text Dataset

```python
documents = [
    'I loved the movie. It was fantastic and thrilling!',
    'The movie was boring and too long.',
    'What an amazing performance by the actors.',
    'The plot was dull and the characters were predictable.'
]
```

## Tokenization

```python
tokens = []
for doc in documents:
    tokens.extend(word_tokenize(doc))
print(tokens)
```

## Stopwords Removal

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [w.lower() for w in tokens if w.lower() not in stop_words and w not in string.punctuation]
print(filtered_tokens)
```

## Stemming and Lemmatization

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stemmed = [stemmer.stem(w) for w in filtered_tokens]
lemmatized = [lemmatizer.lemmatize(w) for w in filtered_tokens]
print('Stemmed:', stemmed)
print('Lemmatized:', lemmatized)
```

## POS Tagging

```python
print(pos_tag(filtered_tokens))
```

## Feature Engineering

```python
# Bag of Words
bow = CountVectorizer()
print(bow.fit_transform(documents).toarray())
```

```python
# One Hot Encoding
onehot = CountVectorizer(binary=True)
print(onehot.fit_transform(documents).toarray())
```

```python
# TF-IDF
tfidf = TfidfVectorizer()
print(tfidf.fit_transform(documents).toarray())
```

```python
# N-grams
print('Unigrams:', CountVectorizer(ngram_range=(1,1)).fit(documents).get_feature_names_out())
print('Bigrams:', CountVectorizer(ngram_range=(2,2)).fit(documents).get_feature_names_out())
```
